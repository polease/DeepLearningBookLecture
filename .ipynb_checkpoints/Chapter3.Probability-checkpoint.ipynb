# Probability and Information Theory



## Probability Mass Function (PMF) 


> <img src="images/3.Probability/Dice.png" height="30px" width="60px" class="hint" style="text-align:left"/> The domain of X contains discrete values

* The domain of P must be the set of all possible states of x.
 ∀x ∈ x,0≤ P(x)≤1.
* An impossible event has probability 0, and no statecan be less probable than that. Likewise, an event that is guaranteed tohappen has probability 1, and no state can have a greater chance of occurring.•
* x∈xP(x) = 1. We refer to this property as beingnormalized. 

![PMF](images/PMF.png)


Example : Uniformed distribution

$$ P(X = x_i) = \frac{1}{k} \tag{3.1} $$

![Uniformed Discrete](images\3.Probability\UniformDiscrete.png)

## Probability Density Function (PDF)


 * The domain of p must be the set of all possible states of x.
 * ∀x ∈ x, p(x) ≥ 0. Note that we do not require p(x) ≤ 1.
 * $\int p(x)dx = 1 $

Example: Normal Distribution

![Normal Distribution](images/3.Probability/PDF.png)

## Joint, Marginal and Conditional Probability

P(X = x,Y = y) is the joint distribution of X and Y, while P(X = x|Y = y) is the conditional distribution of X given Y
$$
P(X=x) = \Sigma_y P(X=x,Y=y) = \Sigma_y (X=x | Y = y)P(Y=y)
$$


$$
p(x) = \int p(x,y)dy
$$

Following image illustrate the three proability distribution via a two-way table. Highly recommend see [this great video](https://www.youtube.com/watch?v=SrEmzdOT65s) from zedstatistics .

![Probability Distribution](Images/3.Probability/ProbabilityDistribution.png)



```python
import pandas as pd
import numpy as np

tv = pd.read_csv("TVShowProbability.csv",index_col=0) 
tv
```

```python
# Create Marginal Total (column and row)
tv["Total"] = tv.sum(axis=1)
total_row = pd.Series(tv.sum(axis=0))
total_row.name = "RowTotal"
tv = tv.append(total_row)
tv
```

```python
# Convert to two-way frequency probablity table
# P(Gender) = Sum of P(TVShow,Gender) over TVShow
# P(TVShow) = Sum of P(TVShow,Gender) over Gender
tv_prob = tv/tv.ix["RowTotal","Total"]
tv_prob
```

```python
# Add Conditional column over "Female" 
# P(TVShow | Gender) = P(TVShow,Gender) / P(Gender)
tv_prob["Condition | Female"] = tv_prob["Female"] / tv_prob.ix["RowTotal","Female"]
tv_prob
```

## Probablity Calculation



### Chain Rule

> Think joint probability as logic AND operation

$$ P(x^{(1)},...,x^{(n)}) = P(x^{(1)})\prod_{i=2}^{n} P(x^{(i}|x^{(1)},...,x^{(i-1})) $$

$$ P(a,b,c) = P(a|b,c)P(b,c) = P(a|b,c)P(b|c)P(c) $$



```python
# P1(Game of Thrones, Female) = P2 (Game of Thrones | Female) P3(Female)
P1 = tv_prob.ix["Game of Thrones","Female"]
P2 = tv_prob.ix["Game of Thrones", "Condition | Female"]
P3 = tv_prob.ix["RowTotal", "Female"]
print(P1, P2*P3)
```

### Independence

$$\forall x \in X, y \in Y, p(X=x,Y=y) = p(X=x)P(Y=y) $$

$$ X \bot Y $$

```python
# P1(Game of Thrones, Female)
# P2(Game of Thrones)
# P3(Female)
P1 = tv_prob.ix["Game of Thrones","Female"]
P2 = tv_prob.ix["Game of Thrones", "Total"]
P3 = tv_prob.ix["RowTotal", "Female"]
print(P1, P2*P3)
# P1 <> P2* P3, so TVShow and Gender are not independent
```

## Expectation

$$ \mathbb{E}_{x~P}[f(x)]=\Sigma_x P(x)f(x) \tag{3.9} $$

$$ \mathbb{E}_{x~P}[f(x)]=\int_x P(x)f(x)dx \tag{3.10} $$


$$ \mathbb{E}_{x}[\alpha f(x)+\beta g(x)]=\alpha\mathbb{E}_{x}[f(x)] + \beta\mathbb{E}_{x}[g(x)] \tag{3.11} $$

In uniformed distribution, $\mathbb{E}$ become the mean of X.

$$ \mathbb{E}(x) = \frac{1}{N}\sum\limits_i^{n} x_i $$

```python
from sklearn import datasets
boston = datasets.load_boston()
boston_df = pd.DataFrame(boston.data)
boston_df.columns = boston.feature_names
boston_df["PRICE"] = boston.target
boston_df.head()
```

```python
# Expectation of House Price
E_Price = boston_df["PRICE"].mean()
print(E_Price)
```

### Variance and Covariance

$$ \mu = \mathbb{E}(x) $$

$$ Var(x) = (x - \mathbb{E}(x))^2 $$

> Variance measures how far a set of (random) numbers are spread out (Scattered) from their expected value

$$ Cov(f(x),g(y)) = \mathbb{E}\bigg((f(x) - \mathbb{E}(x))(g(y)- \mathbb{E}(g(y)) \bigg)  $$

In uniform distribution, the coveriance becomes below
$$ Cov(X, Y) = \frac{1}{{N}} \sum\limits_{i}^{N}{(X_i – \mu_x)(Y_i – \mu_y)} $$

> Covariance is a measure how **linearly related** two variables are

<a href="http://stats.seandolinar.com/covariance-different-ways-to-explain/"><img src="images/3.Probability/Covariance.png" width="300px" /></a>


**Covariance Matrix**

$$ Cov(x)_{i,j} = Cov(x_i,y_j) $$

$$ Cov(x,x) = Var(x) $$




```python
cov = boston_df.cov()
cov
```

![Covariance Matrix](images/3.probability/CovarianceMatrix.jpeg)
 

```python
boston_df_desc = boston_df.describe()
boston_df_desc
```

```python
# we can see that the square of σ(standard deviation) is equal to Cov(x,x) and Var(x)
boston_df_desc.ix["std","CRIM"]**2, cov.ix["CRIM","CRIM"]
```

```python
# get expectation of boston dataframe features
boston_df_E = boston_df.mean(axis=0)
print(boston_df_E)
```

```python
# set (x - E(x)) to X_hat
boston_df_hat = boston_df - boston_df_E
boston_df_hat.transpose().dot(boston_df_hat)/boston_df_hat.shape[0]
```

## Common Distribution

### Bernoulli Distribution

$$ P(x=1) = \phi \tag{3.16} $$
$$ P(x=0) = 1 - \phi \tag{3.17} $$
$$ P(x=x) = \phi^x (1-\phi)^{1-x} \tag{3.18} $$
$$ \mathbb{E}[x] = \phi \tag{3.19} $$
$$ Var(x) = \phi(1-\phi) \tag{3.20} $$




## Gaussian Distribution

$$  \mathcal{N}(x;\mu,\sigma^2) = \sqrt{\frac{1}{2\pi\sigma^2}}exp\big(-\frac{1}{2\sigma^2}(x-\mu)^2\big) \tag{3.21} $$

```python
mu, sigma = 0, 0.5

sample = np.random.normal(mu,sigma,10000)
hist, edges = np.histogram(sample, density=True, bins=1000)

x = np.linspace(-2,2,1000)
pdf = 1 / np.sqrt(2*np.pi*(sigma**2)) * np.exp(-1/(2*(sigma**2))*((x-mu)**2))

x[0:2], pdf[:2]
```
